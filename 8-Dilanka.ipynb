```python
class Node:
    def __init__(self, val):
        self.l = None
        self.r = None
        self.v = val

class Tree:
    def __init__(self):
        self.root = None

    def getRoot(self):
        return self.root

    def add(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._add(val, self.root)

    def _add(self, val, node):
        if val < node.v:
            if node.l is not None:
                self._add(val, node.l)
            else:
                node.l = Node(val)
        else:
            if node.r is not None:
                self._add(val, node.r)
            else:
                node.r = Node(val)

    def find(self, val):
        if self.root is not None:
            return self._find(val, self.root)
        else:
            return None

    def _find(self, val, node):
        if val == node.v:
            return node
        elif (val < node.v and node.l is not None):
            return self._find(val, node.l)
        elif (val > node.v and node.r is not None):
            return self._find(val, node.r)

    def deleteTree(self):
        self.root = None

    def printTree(self):
        if self.root is not None:
            self._printTree(self.root)

    def _printTree(self, node):
        if node is not None:
            self._printTree(node.l)
            print(str(node.v) + ' ')
            self._printTree(node.r)
            
    def getParent(self, root, node):
        if root.l is not None :
            if root.l.v == node.v :
                return root
            else :
                return self.getParent(root.l , node)
        elif root.r is not None :
            if root.r.v == node.v :
                return root
            else :
                return self.getParent(root.r , node)

            
    def successor(self,root, n): 
        if n.r is not None: 
            return tree.minValue(n.r) 

        p = tree.getParent(self.root,n)
        while( p is not None): 
            if n != p.r : 
                break 
            n = p  
            p = tree.getParent(self.root,n) 
            
        return p
            

    def minValue(self, node): 
        current_node = node 
        while(current_node is not None): 
            if current_node.l is None: 
                break
            current_node = current_node.l 

        return current_node 

tree = Tree()
tree.add(3)
tree.add(4)
tree.add(0)
tree.add(8)
tree.add(2)
tree.printTree()

# finding successor
x = tree.find(2)
successor = tree.successor(tree, x)
if successor == None :
    print ("%d does not have a successor " %(x.v))
else :
    print ("Successor of %d is %d " %(x.v , successor.v))

# print(tree.find(3).v)
# print(tree.find(10))
tree.deleteTree()
tree.printTree()
```
