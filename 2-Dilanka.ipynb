```python
import numpy as np
import matplotlib.pyplot as plt

def plotGraphs(x) :
    y_1 = np.ones(100)
    y_n = x
    y_n_squared = x**2
    y_n_sqrt = x**0.5
    y_nlogn = x*np.log(x)
    y_n_cube = x**3
    with np.errstate(over='ignore'):
        y_2_to_nth_power = 2**x

    plot = plt.figure()
    p = plot.add_subplot()
    p.set_yscale('log')
    p.set_xscale('log')
    p.set_xlim(1,10**15)
    p.set_ylim(1,10**19)
    p.plot(x,y_1,label='1')
    p.plot(x,y_n,label='n')
    p.plot(x,y_n_squared, label='n^2')
    p.plot(x,y_n_sqrt, label='sqrt(n)')
    p.plot(x,y_nlogn, label='nlogn')
    p.plot(x,y_n_cube, label='n^3')
    p.plot(x,y_2_to_nth_power, label='2^n')
    plt.xlabel("x")
    plt.title("Functions")
    plt.legend()

plotGraphs(np.logspace(0,15,100,dtype=np.longfloat))
```
